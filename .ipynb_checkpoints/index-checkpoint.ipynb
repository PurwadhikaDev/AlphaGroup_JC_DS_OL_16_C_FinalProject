```python
orders = pd.read_csv('olist_orders_dataset.csv', parse_dates=['order_purchase_timestamp'])
order_items = pd.read_csv('olist_order_items_dataset.csv')

# Tampilkan informasi dasar dari dataset
print("Ukuran orders:", orders.shape)
print("Ukuran order_items:", order_items.shape)

# --- 2. Hitung Recency & Frequency ---
# Tentukan reference_date sebagai tanggal pembelian terakhir dari seluruh data
reference_date = orders['order_purchase_timestamp'].max()
print("Reference Date:", reference_date)

# Kelompokkan data orders untuk menghitung:
# Recency: jumlah hari sejak transaksi terakhir tiap pelanggan
# Frequency: jumlah order unik tiap pelanggan
rfm = orders.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (reference_date - x.max()).days,
    'order_id': 'nunique'
}).rename(columns={
    'order_purchase_timestamp': 'recency', 
    'order_id': 'frequency'
}).reset_index()

# --- 3. Hitung Monetary ---
# Gabungkan orders dengan order_items untuk mendapatkan total harga tiap order
orders_items = pd.merge(orders[['order_id', 'customer_id']],
                        order_items[['order_id', 'price']],
                        on='order_id', how='left')
# Total pengeluaran per pelanggan (monetary)
monetary = orders_items.groupby('customer_id').agg({'price': 'sum'}) \
                        .rename(columns={'price': 'monetary'}) \
                        .reset_index()

# Gabungkan metrik RFM dengan nilai monetary
rfm = pd.merge(rfm, monetary, on='customer_id', how='left')

print("\nContoh perhitungan awal RFM:")
print(rfm.head())

# --- 4. Berikan Skor RFM ---
# Kita akan membagi tiap metrik ke dalam 4 kuartil (skor 1 s.d. 4)
# Untuk recency: skor lebih tinggi diberikan jika nilai recency rendah (transaksi terbaru)
rfm['R_score'] = pd.qcut(rfm['recency'], 4, labels=[4, 3, 2, 1]).astype(int)
# Untuk frequency: semakin tinggi jumlah order, semakin tinggi skornya
rfm['F_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4]).astype(int)
# Untuk monetary: semakin tinggi total pembelian, semakin tinggi skornya
rfm['M_score'] = pd.qcut(rfm['monetary'].rank(method='first'), 4, labels=[1, 2, 3, 4]).astype(int)

# Total skor RFM (rentang 3-12)
rfm['RFM_score'] = rfm['R_score'] + rfm['F_score'] + rfm['M_score']

print("\nHasil akhir RFM scoring:")
print(rfm.head())
```

```python
import pandas as pd

# --- 1. Muat Data ---
# Pastikan file CSV tersedia di direktori kerja
orders = pd.read_csv('olist_orders_dataset.csv', parse_dates=['order_purchase_timestamp'])
order_items = pd.read_csv('olist_order_items_dataset.csv')

# --- 2. Hitung Recency & Frequency ---
# Tentukan reference_date sebagai tanggal pembelian terakhir dalam data
reference_date = orders['order_purchase_timestamp'].max()

# Kelompokkan data orders untuk menghitung:
# - Recency: jumlah hari sejak transaksi terakhir tiap customer
# - Frequency: jumlah order unik tiap customer
rfm = orders.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (reference_date - x.max()).days,
    'order_id': 'nunique'
}).rename(columns={
    'order_purchase_timestamp': 'recency', 
    'order_id': 'frequency'
}).reset_index()

# --- 3. Hitung Monetary ---
# Gabungkan orders dengan order_items untuk mendapatkan nilai order
orders_items = pd.merge(
    orders[['order_id', 'customer_id']],
    order_items[['order_id', 'price']],
    on='order_id', 
    how='left'
)
# Total pengeluaran per customer (monetary)
monetary = orders_items.groupby('customer_id').agg({'price': 'sum'}) \
                        .rename(columns={'price': 'monetary'}) \
                        .reset_index()

# Gabungkan nilai Monetary ke data RFM
rfm = pd.merge(rfm, monetary, on='customer_id', how='left')

# --- 4. Tentukan Threshold untuk Setiap Metrik ---
# Misalnya, gunakan threshold 180 hari untuk recency,
# dan gunakan median untuk frequency dan monetary.
threshold_recency = 180
median_frequency = rfm['frequency'].median()
median_monetary = rfm['monetary'].median()

print("Threshold recency (hari):", threshold_recency)
print("Median frequency:", median_frequency)
print("Median monetary:", median_monetary)

# --- 5. Definisikan Churn Berdasarkan RFM ---
# Option 1: Pendekatan ketat, pelanggan dianggap churn jika ketiga kondisi terpenuhi.
rfm['churn_strict'] = (
    (rfm['recency'] > threshold_recency) &
    (rfm['frequency'] < median_frequency) &
    (rfm['monetary'] < median_monetary)
).astype(int)

# Option 2: Pendekatan fleksibel "2 dari 3"
# Jika dua atau tiga kondisi terpenuhi, pelanggan dianggap churn.
rfm['churn_flexible'] = (
    ((rfm['recency'] > threshold_recency).astype(int) +
     (rfm['frequency'] < median_frequency).astype(int) +
     (rfm['monetary'] < median_monetary).astype(int)) >= 2
).astype(int)

# Tampilkan contoh hasil RFM beserta label churn
print(rfm.head())

print("\nJumlah customer churn (strict):", rfm['churn_strict'].sum())
print("Jumlah customer churn (flexible):", rfm['churn_flexible'].sum())
print("Jumlah customer ", rfm['monetary'].sum())
```

```python
import pandas as pd
import numpy as np

# ---------------------------------------------------------
# 1. Muat Data
# ---------------------------------------------------------
orders = pd.read_csv('olist_orders_dataset.csv', parse_dates=['order_purchase_timestamp'])

# Pastikan kolom 'order_purchase_timestamp' sudah menjadi datetime
print(orders.info())

# ---------------------------------------------------------
# 2. Identifikasi One-Time vs Repeat Buyer
# ---------------------------------------------------------
# Hitung jumlah order per customer
order_counts = orders.groupby('customer_id')['order_id'].nunique()

# Kumpulkan daftar customer one-time vs repeat
one_time_buyers = order_counts[order_counts == 1].index
repeat_buyers = order_counts[order_counts > 1].index

print(f"Jumlah one-time buyer  : {len(one_time_buyers)}")
print(f"Jumlah repeat buyer    : {len(repeat_buyers)}")

# ---------------------------------------------------------
# 3. Hitung Interpurchase Time untuk Repeat Buyer
# ---------------------------------------------------------
# Sort data agar bisa menghitung selisih hari antar order
orders_sorted = orders.sort_values(['customer_id', 'order_purchase_timestamp'])

# Buat kolom untuk tanggal order sebelumnya (per customer)
orders_sorted['prev_order_date'] = orders_sorted.groupby('customer_id')['order_purchase_timestamp'].shift(1)

# Hitung selisih hari antar order
orders_sorted['days_diff'] = (orders_sorted['order_purchase_timestamp'] - orders_sorted['prev_order_date']).dt.days

# Ambil hanya repeat buyer untuk mencari median interpurchase time
orders_repeat = orders_sorted[orders_sorted['customer_id'].isin(repeat_buyers)].dropna(subset=['days_diff'])
median_interpurchase = orders_repeat['days_diff'].median()
print(f"Median interpurchase time (repeat buyer): {median_interpurchase:.2f} hari")

# Tentukan threshold (contoh: 3x median)
churn_threshold_repeat = 3 * median_interpurchase
print(f"Threshold churn repeat buyer (hari)     : {churn_threshold_repeat:.2f}")

# ---------------------------------------------------------
# 4. Hitung Recency dan Days Since First Purchase
# ---------------------------------------------------------
# Dapatkan tanggal paling akhir di dataset (untuk reference)
max_date = orders['order_purchase_timestamp'].max()

# Dapatkan last_order_date dan first_order_date tiap customer
agg_dates = orders.groupby('customer_id')['order_purchase_timestamp'].agg(['min','max']).reset_index()
agg_dates.columns = ['customer_id','first_order_date','last_order_date']

# Hitung recency (hari sejak order terakhir)
agg_dates['recency'] = (max_date - agg_dates['last_order_date']).dt.days

# Hitung days_since_first (hari sejak order pertama)
agg_dates['days_since_first'] = (max_date - agg_dates['first_order_date']).dt.days

# ---------------------------------------------------------
# 5. Definisikan Churn Berbeda untuk One-Time vs Repeat
# ---------------------------------------------------------
def define_churn(row):
    # Jika repeat buyer
    if row['customer_id'] in repeat_buyers:
        # Churn jika recency > threshold
        return 1 if row['recency'] > churn_threshold_repeat else 0
    else:
        # Untuk one-time buyer: churn jika belum beli kedua kali
        # dan days_since_first > threshold
        return 1 if row['days_since_first'] > churn_threshold_repeat else 0

agg_dates['churn'] = agg_dates.apply(define_churn, axis=1)

# ---------------------------------------------------------
# 6. Lihat Hasil
# ---------------------------------------------------------
print("\nContoh data hasil (customer_id, first_order_date, last_order_date, recency, days_since_first, churn):")
print(agg_dates.head(10))

churn_rate = agg_dates['churn'].mean() * 100
print(f"\nChurn Rate (persen): {churn_rate:.2f}%")
```

```python
import pandas as pd

# --------------------------------------------------------
# 1. Load Dataset
# --------------------------------------------------------
orders = pd.read_csv('olist_orders_dataset.csv', parse_dates=['order_purchase_timestamp'])
customers = pd.read_csv('olist_customers_dataset.csv')

# Pastikan kolom tanggal sudah bertipe datetime
print("Info orders:")
print(orders.info())
print("\nInfo customers:")
print(customers.info())

# --------------------------------------------------------
# 2. Merge Data on 'customer_id'
#    -> Sehingga kita dapat 'customer_unique_id' dari customers
# --------------------------------------------------------
df_merged = pd.merge(orders, customers, on='customer_id', how='left')

# --------------------------------------------------------
# 3. Hitung order_count, first_order_date, dan last_order_date per customer_unique_id
# --------------------------------------------------------
agg_data = df_merged.groupby('customer_unique_id').agg({
    'order_id': 'nunique',  # jumlah pesanan unik
    'order_purchase_timestamp': ['min', 'max']  # tanggal order pertama & terakhir
}).reset_index()

# Rapikan nama kolom
agg_data.columns = ['customer_unique_id', 'order_count', 'first_order_date', 'last_order_date']

# --------------------------------------------------------
# 4. Tambahkan kolom recency dan days_since_first
#    -> recency: hari sejak pembelian terakhir
#    -> days_since_first: hari sejak pembelian pertama
# --------------------------------------------------------
# Tentukan reference_date sebagai tanggal order paling akhir di keseluruhan data
max_date = df_merged['order_purchase_timestamp'].max()

agg_data['recency'] = (max_date - agg_data['last_order_date']).dt.days
agg_data['days_since_first'] = (max_date - agg_data['first_order_date']).dt.days

# --------------------------------------------------------
# 5. Tandai One-Time vs Repeat Buyer
# --------------------------------------------------------
agg_data['buyer_type'] = agg_data['order_count'].apply(lambda x: 'one_time' if x == 1 else 'repeat')

# --------------------------------------------------------
# 6. Definisikan threshold dan logic churn
#    (Anda bisa ubah 180 menjadi nilai lain sesuai kebutuhan)
# --------------------------------------------------------
threshold_one_time = 180
threshold_repeat = 180

def define_churn(row):
    if row['buyer_type'] == 'one_time':
        # One-time buyer: churn jika melewati X hari tanpa order kedua
        # (di sini kita pakai days_since_first)
        return 1 if row['days_since_first'] > threshold_one_time else 0
    else:
        # Repeat buyer: churn jika recency > X hari
        return 1 if row['recency'] > threshold_repeat else 0

agg_data['churn'] = agg_data.apply(define_churn, axis=1)

# --------------------------------------------------------
# 7. Lihat Hasil
# --------------------------------------------------------
print("\nContoh Data Hasil:")
print(agg_data.head(10))

# Hitung churn rate secara keseluruhan
churn_rate = agg_data['churn'].mean() * 100
print(f"\nChurn Rate (total): {churn_rate:.2f}%")

# Pisahkan churn rate one-time vs repeat
one_time_data = agg_data[agg_data['buyer_type'] == 'one_time']
repeat_data = agg_data[agg_data['buyer_type'] == 'repeat']

churn_rate_one_time = one_time_data['churn'].mean() * 100 if len(one_time_data) > 0 else 0
churn_rate_repeat = repeat_data['churn'].mean() * 100 if len(repeat_data) > 0 else 0

print(f"Churn Rate (one-time buyers): {churn_rate_one_time:.2f}%")
print(f"Churn Rate (repeat buyers)  : {churn_rate_repeat:.2f}%")
```
